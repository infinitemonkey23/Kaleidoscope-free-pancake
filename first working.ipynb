```python
from PIL import Image, ImageDraw, ImageOps
import numpy as np
import gizeh as gz
from matplotlib.image import imread
%matplotlib inline
from matplotlib import pyplot as plt

myimg = imread('2.jpg')
h,w ,other= myimg.shape


center=  np.array([(13*w/17), h*0.53])
user_rota=7*np.pi/3
def my_filter(_image):
    radi=200
    
    surface = gz.Surface(692,300)
    fill = (gz.ImagePattern(_image, pixel_zero=center,extend="reflect")
            .scale(1,1, center=center))
    for (nfaces,angle,rot,pangle,flip) in ([3, np.pi/3, np.pi/2,-np.pi/3,0],
                                           [3, 0, np.pi/6,0,1],[3, -np.pi/3, -np.pi/6,np.pi/3,0]):
        
        xy = (346+(np.sin(pangle)*radi),0+(np.cos(pangle)*radi))
        shape = gz.regular_polygon(radi,nfaces,angle=rot, xy= xy,
                fill=(fill.rotate(user_rota, center)).scale(1) )

  #      shape=gz.polyline(points=)
        if flip ==1:
        
            shape=shape.scale(-1,1,center=xy)
        shape.draw(surface)
        lopper= surface.get_npimage()
#        if flip ==1:
#            fliplopper= lopper[:,::-1]
        
    newIm = Image.fromarray(lopper, "RGB")
#    newIm.show()
# ####   newIm.save("out.jpg")
    return(newIm)

trapez=my_filter(myimg)
#(.5+ .05*np.sin(2*np.pi)))
#-np.pi/3#
#xy = (400+(np.sin(rot)*offs), 250+(np.cos(rot)*offs))





w,h = trapez.size
canvas=Image.new("RGB",(w,h*2),color=255)
canvas.paste(trapez,(0,h))

flipped=ImageOps.flip(trapez)
canvas.paste(flipped,(0,0))
#canvas.show()
#canvas.save("kal6.jpg")


def hexaganalator(_image):
    w,h= _image.size
    hradius=2*h/3
    wradius=w/2
    my_canvas=gz.Surface(w*2,h*2)
    leftvert=(w/4)
    rightvert=(w*3/4)
    
    _image=np.array(_image)
    center=  np.array([(w/2), h/2])
    pattern=(gz.ImagePattern(_image,pixel_zero=(0,0)))
    hexs=[(w/2,h/2),(w/2,-h/2),(w/2,3*h/2),(-w/4,0),(5*w/4,0),(-w/4,h),(5*w/4,h)      ]
    for _sw, _sh in hexs:
        hex_pat = gz.polyline(points=[(leftvert,0), (rightvert,0), (w,h/2), (rightvert,h),
                                    (leftvert,h),(0,h/2),(leftvert,0)], stroke_width=0, fill=pattern,xy=(_sw,_sh))
        hex_pat.draw(my_canvas)
        outp= my_canvas.get_npimage()
    
    
    
    newIm = Image.fromarray(outp, "RGB")
    newIm.show()
#    newIm.save("fullpattern14.jpg")
hexaganalator(canvas)
```

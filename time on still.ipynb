```python
''' 
A program to create Kaleidoscopic images or videos from imported images or videos.
'''
import moviepy.editor as mpy
import moviepy
from PIL import Image, ImageDraw, ImageOps
import numpy as np
import gizeh as gz
from matplotlib.image import imread
from io import BytesIO


class Stillkaleido:
    """
    Class representation of the function for creating a jpg kaleidoscope version of an image.
    """
    
    
    def __init__(self,openedim):
        '''
        initialize the file into a moviepy video.
        '''
        
        self.theimage = np.array(openedim)    
        
        self.radi=200
        self.zoom=1
        self.triangletype=1
        self.animation=False
        
        
    def make_image(self,rotated):        
            
        if rotated==False:
            self.rotation=(np.pi/6)
        else:
            self.rotation=(np.pi/6)+rotated
            
        if self.triangletype==1:    
            self.Equilateraltrianglize()
            
        elif self.triangletype==30:
            self.Half_equilateral_t() 
            
        elif self.triangletype==245:
            self.two_45_sides()
        elif self.triangletype==260:     
            self.twomirror60()
            
        self.Trapeezer()
        
        if self.triangletype==245:
            self.Rect_tesselator()
        else:
            self.Hexaganalator()    
        return(self.tring)   
    
    
    def set_height_of_sub_im(self):
        """
        Changes the height of the image to be sected by user, for use in GUI
        """
        
        self.height_of_sub=height_of_sub
    
    
    def setisos(self):
        
        self.triangletype=1
    
    
    def set30degrees(self):
        
        self.triangletype=30
        
    def settwomirror60(self):
        
        self.triangletype=260
    
    
    def setpos(self,center):
        self.center=center
    
    def set2x45_tiranle(self):
        self.triangletype=245
        
    def set_zoom(self,level):    
    
        self.zoom=level
        
    
    def twomirror60(self):
        """
        first step in creating a 3 mirror, 12 fold symetry pattern.
        """
        self.radi=self.radi*4
        self.centre=np.array([self.center])
        self.hori=(int(np.trunc((np.cos(np.pi/6))*(3*self.radi/2))))
        self.vert=int(self.radi*1.5)

        surface=gz.Surface(int(4*self.hori/2),int(3*self.vert/4),bg_color=None)
        fill=(gz.ImagePattern(self.theimage, pixel_zero=self.center,extend="reflect").scale(self.zoom,center=self.center)) 
        for (nfaces,angle,rot,pangle,flip) in ([3, np.pi/3, np.pi/2,-np.pi/3,0],
                                               [3, 0, np.pi/6,0,1],[3, -np.pi/3, -np.pi/6,np.pi/3,0]):
            xy = ((self.hori+(np.sin(pangle)*self.radi)),(np.cos(pangle)*self.radi))
            shape = gz.regular_polygon(self.radi+1,nfaces,angle=rot, xy= xy,
                                       fill=(fill.rotate(self.rotation, self.center)).scale(1) )
            if flip ==1:
                shape=shape.scale(-1,1,center=xy)
           
            shape.draw(surface)  

        self.tring=Image.fromarray(surface.get_npimage(), "RGB")


        
    def two_45_sides(self):
        """
        first step in creating an Isosceles right triangle pattern.
        """

        self.centre=np.array([self.center])
        self.hori=int(1.5*self.radi)
        self.vert=int(1.5*self.radi)
    #    mycenter=(self.center[0],self.center[1])
        middle=((self.center[0]-(4*self.hori/5)),(self.center[1]-(2*self.vert/3)))
        
        
        surface=gz.Surface((2*self.hori),(self.vert),bg_color=None)
        fill=(gz.ImagePattern(self.theimage, pixel_zero=middle,extend="reflect").scale(self.zoom,center=self.center)) 
        verticies=[(0,-1),(1,-1),(self.hori,self.vert-1),(self.hori,self.vert),(0,self.vert)]
        coverp=[(0,-1),(1,-1),(self.hori+1,self.vert-1),(self.hori+1,self.vert),(0,self.vert)]

        orig = gz.polyline(points=verticies,fill=(fill.rotate(self.rotation,self.center)))
        mirro= gz.polyline(points=verticies,fill=(fill.rotate(self.rotation,self.center)))

        mirro=mirro.scale(-1,1,center=((self.hori),0))
        mirro=mirro.rotate(-np.pi/2,center=(self.hori,self.vert))
    #    mirro=mirro.translate((0,1))
        set1=gz.Group([mirro,orig])               
        set2=gz.Group([mirro,orig])
        set2=set2.scale(-1,1,center=((self.hori),0))
     #   set3=gz.Group([mirro,orig])
      #  set2=set2.rotate(np.pi/3,center=((self.hori*2),0))
      #  set3=set3.rotate(-np.pi/3,center=((self.hori*2),0))
     #   set2=set2.translate((0,0))
      #  set3=set3.translate((0,0))
    #    cover.draw(surface)
        set1.draw(surface)
        set2.draw(surface)
    #    set3.draw(surface)
   #     mirro.draw(surface)
   #     orig.draw(surface)
        
    
        self.tring=Image.fromarray(surface.get_npimage(), "RGB")
     
    
    def Rect_tesselator(self):
        
        maincanvas=gz.Surface(4*self.hori,4*self.vert,bg_color=None)

        _image=np.array(self.tring)
   #     center=np.array([self.vert, self.hori])
        
        pattern=(gz.ImagePattern(_image,pixel_zero=(0,0)))
        hexs=[(-self.hori,-self.vert),(self.hori,-self.vert),(3*self.hori,-self.vert),
                 (-self.hori,self.vert),(self.hori,self.vert),(3*self.hori,self.vert),
                 (-self.hori,3*self.vert),(self.hori,3*self.vert),(3*self.hori,3*self.vert)]
        for _sw, _sh in hexs:
            hex_pat = gz.square(l=self.hori*4, stroke_width=0, 
                                      fill=pattern,xy=(_sw,_sh))
            hex_pat.draw(maincanvas)
            outp= maincanvas.get_npimage()
        
        self.tring = Image.fromarray(outp, "RGB")
        
        if self.animation==True:
            
            return(self.convertToJpeg(self.tring))
        
        
        
        
    def Half_equilateral_t(self):
        """
        first step in creating a 3 mirror, 12 fold symetry pattern.
        """

        self.centre=np.array([self.center])
        self.hori=(int(np.trunc((np.cos(np.pi/6))*self.radi)))
        self.vert=int(1.5*self.radi)
    #    mycenter=(self.center[0],self.center[1])
        middle=((self.center[0]-(4*self.hori/5)),(self.center[1]-(2*self.vert/3)))
        
        
        surface=gz.Surface((4*self.hori),(self.vert),bg_color=None)
        fill=(gz.ImagePattern(self.theimage, pixel_zero=middle,extend="reflect").scale(self.zoom,center=self.center)) 
        verticies=[(-1,self.vert),(self.hori+1,self.vert),(self.hori+1,0)]
        coverp=[(self.hori-1,self.vert),(self.hori+1,self.vert),(self.hori+1,0),(self.hori-1,0)]
      #  for (flip,rot,offset) in ([0,0,0],[1,0,0]):
        orig = gz.polyline(points=verticies,fill=(fill.rotate(self.rotation,self.center)))
        mirro= gz.polyline(points=verticies,fill=(fill.rotate(self.rotation,self.center)))
        # cover is a little sliver to cover the join between the original image adn mirrored, otherwise
        #when rotated they leave a line
        cover= gz.polyline(points=coverp,fill=(fill.rotate(self.rotation,self.center)))
        orig=orig.translate((self.hori-1,0))
        mirro=mirro.translate((self.hori-1,0))
        cover=cover.translate((self.hori,0))
        mirro=mirro.scale(-1,1,center=((2*self.hori),0))
        
        set1=gz.Group([mirro,orig,cover])               
        set2=gz.Group([mirro,orig,cover])
        set3=gz.Group([mirro,orig,cover])
        set2=set2.rotate(np.pi/3,center=((self.hori*2),0))
        set3=set3.rotate(-np.pi/3,center=((self.hori*2),0))
        set2=set2.translate((0,0))
        set3=set3.translate((0,0))
        cover.draw(surface)
        set1.draw(surface)
        set2.draw(surface)
        set3.draw(surface)
        self.tring=Image.fromarray(surface.get_npimage(), "RGB")
     
        self.hori=self.hori*2
        
    
    
    
    def Equilateraltrianglize(self):
        self.centre=np.array([self.center])
        self.hori=(int(np.trunc((np.cos(np.pi/6))*2*self.radi)))
        self.vert=int(self.radi*1.5)

        surface=gz.Surface((2*self.hori)+2,self.vert,bg_color=None)
        fill=(gz.ImagePattern(self.theimage, pixel_zero=self.center,extend="reflect").scale(self.zoom,center=self.center)) 
        for (nfaces,angle,rot,pangle,flip) in ([3, np.pi/3, np.pi/2,-np.pi/3,0],
                                               [3, 0, np.pi/6,0,1],[3, -np.pi/3, -np.pi/6,np.pi/3,0]):
            xy = (1+(self.hori+(np.sin(pangle)*self.radi)),0+(np.cos(pangle)*self.radi))
            shape = gz.regular_polygon(self.radi+1,nfaces,angle=rot, xy= xy,
                                       fill=(fill.rotate(self.rotation, self.center)).scale(1) )
            if flip ==1:
                shape=shape.scale(-1,1,center=xy)
           
            shape.draw(surface)  

        self.tring=Image.fromarray(surface.get_npimage(), "RGB")
     

    def timerot(self,time):
        self.rotation=self.startrot+(time*self.rate)
        
    
    
    def Trapeezer(self):
        if self.triangletype==245:
            mirrorcanva=Image.new("RGB",((2*self.hori),2*self.vert),color=(155,155,155))
        else:    
            mirrorcanva=Image.new("RGB",((2*self.hori)+2,2*self.vert),color=(155,155,155))
        mirrorcanva.paste(self.tring,(0,self.vert))
        flipped=ImageOps.flip(self.tring)
        mirrorcanva.paste(flipped,(0,0))
        
        self.tring=mirrorcanva
    
    def Hexaganalator(self):

        maincanvas=gz.Surface((4*self.hori)-10,4*self.vert,bg_color=None)
        leftvert=((self.hori/2))
        rightvert=((3*self.hori/2))
        _image=np.array(self.tring)
   #     center=np.array([self.vert, self.hori])
        
        pattern=(gz.ImagePattern(_image,pixel_zero=(0,0)))
        hexs=[(4-self.hori/2,2*self.vert),(4-self.hori/2,0),(self.hori,self.vert),
                (self.hori,-self.vert),(self.hori,3*self.vert),((5*self.hori/2)-4,0),
              ((5*self.hori/2)-4,2*self.vert)]
        for _sw, _sh in hexs:
            hex_pat = gz.polyline(points=[(leftvert+1,0), (rightvert-1,0), ((2*self.hori)-2,self.vert), (rightvert-1,2*self.vert),
                                          (leftvert+1,2*self.vert),(2,self.vert),(leftvert+1,0)], stroke_width=0, 
                                      fill=pattern,xy=(_sw,_sh))
            hex_pat.draw(maincanvas)
            outp= maincanvas.get_npimage()
        
        self.tring = Image.fromarray(outp, "RGB")
        
        if self.animation==True:
            
            return(self.convertToJpeg(self.tring))
         #   return(outp)

    
    def saveimage(self,text):
        name=text+".jpg"
        self.tring.save(name)
    
    
    def convertToJpeg(self,im):
        """"
        to overcome masive mp4 file sizes this, sudo_saves the file as a jpeg image before it is
        animated adn is compressed in the process
        """
        
        
        with BytesIO() as f:
            im.save(f, format='JPEG',optimize=True,quality=30)
            f.seek(0)
            ima_jpg = Image.open(f)
            imag=np.array(ima_jpg)
        return  np.array(ima_jpg) 
      #  return ( Image.fromarray(imag, "RGB"))
            
    
    def spinner(self, time):         
             
        self.timerot(time)
        if self.triangletype==1:    
            self.Equilateraltrianglize()
        elif self.triangletype==30:
            self.Half_equilateral_t()     
        elif self.triangletype==245:
            self.two_45_sides()
            
        self.Trapeezer()
        
        if self.triangletype==245:
            return (self.Rect_tesselator())
        else:
            return (self.Hexaganalator())
    
    def runner(self, time):
        """
        this moves postion along with time when making a moving scope
        """
        
        (x,y)=(self.startloc[0]+time*self.xvec,self.startloc[1]+time*self.yvec)
        
        self.setpos((x,y))   
        
        if self.triangletype==1:    
            self.Equilateraltrianglize()
        elif self.triangletype==30:
            self.Half_equilateral_t()     
        elif self.triangletype==245:
            self.two_45_sides()
            
        self.Trapeezer()
        
        if self.triangletype==245:
            return (self.Rect_tesselator())
        else:
            return (self.Hexaganalator())
            
        
    
    
    
    
    def animate_rota(self,D,savename):
        self.rate=(2*np.pi)/(D)
        self.startrot=self.rotation
        self.animation=True 
        
        clip = mpy.VideoClip(self.spinner, duration=D)
        clip.write_videofile(savename, preset="veryslow",threads=4, fps=14,audio=False)
        self.animation=False
        
        #, preset="slower"
        

    def jounrney(self,D,savename,points):
        self.xvec=(points[2]-points[0])/D
        self.yvec=(points[3]-points[1])/D
        self.startloc=(points[0],points[1])
        self.setpos(self.startloc)
        self.animation=True
        clip = mpy.VideoClip(self.runner, duration=D)
    #    clip.preview()
        clip.write_videofile(savename, preset="veryslow",threads=4, fps=14,audio=False)
        self.animation=False
        
        
        
myclip=Image.open("C:/Users/Tom/Documents/Python Scripts/Kaleidoscope/2.jpg")        
        
a=Stillkaleido(myclip)
a.setpos((680,460))
a.settwomirror60()
#a.set30degrees()
b=a.make_image(False)
#b.show()
```

```python
from tkinter.filedialog import askopenfilename,asksaveasfilename
from tkinter import *
from PIL import Image, ImageDraw, ImageOps, ImageTk


class Window(Frame):
    
    def __init__(self,master=None):
        Frame.__init__(self,master)
        self.master=master
        self.init_window()
                  
    def init_window(self):
        self.seek_start_pos=False
        self.seek_end_pos=False
        self.journey=False
        self.val_check=BooleanVar()
        self.started=False
        self.rotated=False
        self.pos=(0,0)
        self.master.title("Keleidoscopia")
        self.pack(fill=BOTH, expand=1)
        self.leftframe=LabelFrame(self.master,text="Sample Image")
        self.leftframe.pack(side=LEFT)
        
        mymenu=Menu(self.master)          
        self.master.config(menu=mymenu)
        
        file=Menu(mymenu)
        file.add_command(label="Open File",command=self.selectfile)
        file.add_command(label="Save Image",command=self.savefunc)
        mymenu.add_cascade(label="File",menu=file)
        
        options=Menu(mymenu)
        options.add_command(label="Isosceles Triangle", command=self.setisos)
        options.add_command(label="30deg Triangle", command=self.set30deg)
        options.add_command(label="Isosceles Right Triangle", command=self.set45_tiranle)
        mymenu.add_cascade(label="Options",menu=options)
        
        
        
        edit=Menu(mymenu)

        edit.add_command(label="Show Text", command=self.showTxt)
  #      edit.add_command(label="Kaleidoscope", command=self.showversion)
        mymenu.add_cascade(label="Edit",menu=edit)
        
        self.master.bind_all("<MouseWheel>", self._on_mousewheel)
        
        
        self.mycanvas = Canvas(self.leftframe,width=400,height=600,bd=2)
        self.mycanvas.grid(row=1,column=0,rowspan=4,columnspan=3,)
 
        
        self.an_rot_button=Button(self.leftframe,text="Rotation Animation" ,command=self.animate_rot)
        self.an_rot_button.grid(row=7,column=3)
        
        
        self.journeycb = Checkbutton(self.leftframe, variable=self.val_check,text="Journey Animation",
                                    onvalue=True, offvalue=False,
                                    command=self.changebutton)
        
        self.journeycb.grid(row=6,column=3)
        
    
        self.slider = Scale(self.leftframe, from_=-2*np.pi, to=2*np.pi, orient=HORIZONTAL, label="Image Rotation Angle",
                            command=self.rotator)
        self.slider.grid(row=7,column=0,columnspan=2)
        

    
        self.zoomer = Scale(self.leftframe, from_=-5, to=5, orient=HORIZONTAL,label="Sample Scale Factor",
                            resolution=0.5, command=self.zoom_changer)
        self.zoomer.grid(row=7,column=2)
        
        
         
    def selectfile(self):
        if self.started==True:
            self.my2canvas.destroy()
        name = askopenfilename(initialdir="C:/Users/Tom/Documents/Python Scripts/Kaleidoscope/",
                           filetypes =(("jpeg files","*.jpg",".png"),("png files","*.png"),("All Files","*.*")),
                           title = "Choose a file.")
        self.filedir=name
        
        
        self.my2canvas = Canvas(self.master,width=1500,height=1500,bd=2,)
        self.my2canvas.pack(side=RIGHT)
        self.openfile()
        
        
    def openfile(self):    
                
        
        self.myclip = Image.open(self.filedir)
        (w,h)= self.myclip.size
        self.ratio=1
        if w>400 or h>600:           
            wratio=400/w        
            hratio=600/h
            self.ratio=wratio
            if wratio>hratio:
                self.ratio=hratio
            nw=int(np.trunc(w*self.ratio))
            nh=int(np.trunc(h*self.ratio))                
                
            clipped=self.myclip.resize((nw,nh))
        
            _intermediate=ImageTk.PhotoImage(clipped)
        else:
            _intermediate=ImageTk.PhotoImage(self.myclip)
        self.myshownIm=_intermediate
        
        
        self.trianglevers=Stillkaleido(self.myclip)
        

        self.mycanvas.create_image(0,0,image=self.myshownIm,anchor="nw")
     #   self.mycanvas.config(scrollregion=(0,0,w,h))     
        self.mycanvas.bind("<Button 1>",self.printcoords)
    
    
    
    def showversion(self):
               
        
        self.imager=ImageTk.PhotoImage(self.trianglevers.make_image(self.rotated))
        
        self.my2canvas.create_image(0,0,image=self.imager,anchor="nw")
        
    def rotator(self,event):

        change=self.slider.get()
        self.rotated=change
        
        self.rot_changer()
        
    def zoom_changer(self,event):
        
        if self.started==True:
            newzoom=self.zoomer.get()
            if newzoom<0:
                newzoom=1/newzoom
            self.trianglevers.set_zoom(newzoom)
            
            self.showversion()
            
    
    def rot_changer(self):
    
        if self.started==True:
            self.showversion()
      #  self.trianglevers=Stillkaleido(self.myclip,self.rotated)
    
            
    def setisos(self):
        
        self.trianglevers.setisos()
        
   
    def set30deg(self):
        
        self.trianglevers.set30degrees()
    
    
    def set45_tiranle(self):
        
        self.trianglevers.set2x45_tiranle()
        

    def showTxt(self):
        text=Label(self,text="Wooop!, there it is!")
        text.pack()
        
        
    def client_exit(self):
        root.destroy()
        
        
    def printcoords(self,event):
        self.started=True
        
        
        x = self.mycanvas.canvasx(event.x)/self.ratio
        y = self.mycanvas.canvasy(event.y)/self.ratio
        self.pos=(int(x),int(y))
        if self.seek_start_pos:
            self.positions[0],self.positions[1]=self.pos[0],self.pos[1]
            self.startlb.destroy()
            self.startlb = Label(self.leftframe, text=(self.positions[0],self.positions[1]))
            self.startlb.grid(row=6,column=1)
            self.seek_start_pos=False
        if self.seek_end_pos:
            self.positions[2],self.positions[3]=self.pos[0],self.pos[1]
            self.endlb.destroy()
            self.endlb = Label(self.leftframe, text=(self.positions[2],self.positions[3]))
            self.endlb.grid(row=6,column=2)
            self.seek_end_pos=False
        self.trianglevers.setpos(self.pos)
        self.showversion()  
        
    
    def _on_mousewheel(self, event):
        
        if event.delta>0:
            if self.rotated<np.pi*2:
                self.rotated+=((event.delta/480))
            else:
                self.rotated=-2*np.pi
        if event.delta<0:
            if self.rotated>=-np.pi*2:
                self.rotated+=((event.delta/480))
            else:
                self.rotated=2*np.pi
        self.rot_changer()
        self.slider.set(self.rotated)
        
        
    def savefunc(self):
        savename = asksaveasfilename(filetypes = (("jpeg files","*.jpg"),("png files",".png")))

        if savename ==None:
            return
        else:       
            self.trianglevers.saveimage(savename)
            
    
    def animate_rot(self):
        savename = asksaveasfilename(defaultextension=".mp4")
        if savename ==None:
            return
        else:
            if self.journey==False:
                print("fuck")
                self.trianglevers.animate_rota(16,savename)
            elif self.journey==True:
                self.trianglevers.jounrney(16,savename,self.positions)
                
                
    def changebutton(self):
        if self.val_check.get()==True:
            self.journey_menu()
        elif self.val_check.get()==False:
            self.journey_close()
    
    
    def journey_menu (self):
        self.journey=True
        self.positions=[5,12,0,0]
        self.startlb = Label(self.leftframe, text=(self.positions[0],self.positions[1]))
        self.startlb.grid(row=6,column=1)
        self.endlb = Label(self.leftframe, text=(self.positions[2],self.positions[3]))
        self.endlb.grid(row=6,column=2)
        self.set_start_b=Button(self.leftframe, text="Set Starting Position",command=self.set_start_pos)
        self.set_start_b.grid(row=5,column=1)
        self.set_end_b=Button(self.leftframe, text="Set End Position",command=self.set_end_pos)
        self.set_end_b.grid(row=5,column=2)
        
    
    def journey_close(self):
        self.jounrney=False
        self.startlb.destroy()
        self.endlb.destroy()
        self.set_end_b.destroy()
        self.set_start_b.destroy()
        
        
    def set_start_pos(self):
        self.seek_start_pos=True
        
    
    def set_end_pos(self):
        self.seek_end_pos=True
        
        
root=Tk()
root.geometry("1400x800")
app=Window(root)
root.mainloop()
    
```
